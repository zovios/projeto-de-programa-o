fazer um algoritimo e depois o programa que recebe 4 notas do aluno e calcula a media aritimeditica e imprimi a media.

ler aluno_1 ler aluno_2 ler aluno_3 ler aluno_4.

```python
# variaveis que operam aritmeticamente o valor da media

#entrada
aluno_1 = 8
aluno_2 = 6
aluno_3 = 7
aluno_4 = 10
```
