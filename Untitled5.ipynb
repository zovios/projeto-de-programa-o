```python
#pergunta 2

numero1 = int(input("digite o numero que voce deseja comparar aqui: "))
numero2 = int(input("agora, digite outro numero: "))

if numero1 < numero2:
  print(f"{numero1} é menor que {numero2}")

elif numero1 > numero2:
  print(f"{numero1} é maior que {numero2}")

else:
  print(f"{numero1} é igual a {numero2}")
```

```python
#pergunta 3

nota1 = int(input("nota do primeiro bimestre: "))
if nota1 < 0 or nota1 > 10:
  print("nota invalida")
  nota1 = float(input("digite novamente sua nota: "))

nota2 = int(input("nota do segundo bimestre: "))
if nota2 < 0 or nota2 > 10:
  print("nota invalida")
  nota1 = float(input("digite novamente sua nota: "))

nota3 = int(input("nota do terceiro bimestre: "))
if nota3 < 0 or nota3 > 10:
  print("nota invalida")
  nota1 = float(input("digite novamente sua nota: "))

nota4 = int(input("nota do quarto bimestre: "))
if nota4 < 0 or nota4 > 10:
  print("nota invalida")
  nota1 = float(input("digite novamente sua nota: "))

nota_final = (nota1 + nota2 + nota3 + nota4) / 4

if nota_final >= 7:
  print("voce foi aprovado")
elif nota_final >= 5:
  print("voce esta de recuperação")
else:
  print("voce foi reprovado")
```

```python
#pergunta 7

print("opção 1 = soma")
print("opção 2 = subtrair")
print("opção 3 = dividir")
print("opção 4 = multiplicar")
print("----------------------")
print("para opção 1, digite 1")
print("para opção 2, digite 2")
print("para opção 3, digite 3")
print("para opção 4, digite 4")

opção = int(input("digite aqui a opção desejada: "))
primeiro_numero = int(input("digite um numero: "))
segundo_numero = int(input("digite o segundo numero da equação"))

if opção == 1:
  print(f"a soma de {primeiro_numero} e {segundo_numero} é igual a {primeiro_numero + segundo_numero}")

elif opção == 2:
  print(f"a subtração de {primeiro_numero} e {segundo_numero} é igual a {primeiro_numero - segundo_numero}")

elif opção == 3:
  print(f"a divisao de {primeiro_numero} e {segundo_numero} é igual a {primeiro_numero / segundo_numero}")

elif opção == 4:
  print(f"a multiplicação de {primeiro_numero} e {segundo_numero} é igual a {primeiro_numero * segundo_numero}")
```
