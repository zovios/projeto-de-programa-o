```python
# Estrutura de decisões (if, elif e else)

# Escreve um programa que compara 02 números

# entradas
a = int (input("digite um numero inteiro"))
b = int (input("digite outro numero inteiro"))

# entradas / sáida

if a == b:
  print (f"{a} é igual a {b}")

else:
  print (f"{a} é diferente de {b}")
```

```python
# escrever um programa que verifica se o numero digitado é par ou impar

# entrada

numero = int (input("digite um numero: "))

# processamento / saida

if numero % 2 == 0:

  print (f"o numero {numero} é par.")

else:

  print (f"o numero {numero} é impar.")
```

```python
# escrever um programa de uma calculadora simples. O usuario ira digitar 02 numeros e escolherá uma operação aritimética

# entrada

print ("--------------CALCULADORA-------------")
print ("                                      ")
print ("-----------------MENU-----------------")
print ("                                      ")
print ("opção = 1 - Soma                      ")
print ("opção = 2 - subtração                 ")
print ("opção = 3 - multplicação              ")
print ("opção = 4 - divisão                   ")
print ("--------------------------------------")

# entrada
numero = float (input("digite um numéro: "))
opcao = int (input ("Escolha a opção desejda: "))
numero1 = float (input("Digite um numero: "))

# processamento / saida

if opcao == 1:
  print (f"A soma de {numero} e {numero1} é igual a {numero + numero1}")

elif opcao == 2:
  print (f"A subtração de {numero} e {numero1} é igual a {numero - numero1}")

elif opcao == 3:
  print (f"A multiplicação de {numero} e {numero1} é igual a {numero * numero1}")

elif opcao == 4:
  print (f"a divisão de {numero} e {numero1} é igual a {numero / numero1}")

else:
  print ("Opção inválida")
```
