```python
# loops - laços e oteraçoes

for i in range (5):
  print(i)
```

```python
# escrever um programa que recebe um numero e retorna somente os numeros pares.

numero = int(input("digite um numero: "))

for i in range(2, numero + 1):
  if numero % 2 == 0:
    print(i)
```

```python
# escrever um programa que recebe um numero e imprimi uma contagem regressiva

numero = int(input("digite um numero: "))

for i in range(numero + 1):
  print(numero - i)
```

```python
palavra = str (input("digite uma palavra: "))
palindromo = ""

for i in (palavra):
  palindromo = i + palindromo
print(palindromo)
```

```python
# escrever um programa que recebe um numero de 1 a 10 e imprimi a tabuada.

numero = int(input("digite um numero de 1 a 10: "))
if numero < 0 or numero > 10:
  print("numero invalido")
  numero = int(input("digite o seu numero de 1 a 10 novamente: "))

for i in range (1, 11):
  print(f"{numero} x {i} = {numero * i} ")
```
